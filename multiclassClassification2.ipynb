```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
_,ax=plt.subplots()
scatter=ax.scatter(iris.data[:,0],iris.data[:,1],c=iris.target)
ax.set(xlabel=iris.feature_names[0],ylabel=iris.feature_names[1])
_=ax.legend(scatter.legend_elements()[0],iris.target_names,loc="lower right",title="classes")
```

```python
_,ax=plt.subplots()
scatter=ax.scatter(iris.data[:,0],iris.data[:,1],c=iris.target)
ax.set(xlabel=iris.feature_names[2],ylabel=iris.feature_names[3])
_=ax.legend(scatter.legend_elements()[0],iris.target_names,loc="lower right",title="classes")
```

```python
iris.target[0:51]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=200)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([iris.data[51]])
```

```python
model.predict([[6.3, 2.7,6.3,2.7]])
```

```python
model.predict([[6.9,3.2,6.9,3.2]])
```

```python
y_predicted=model.predict(x_test)
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```
